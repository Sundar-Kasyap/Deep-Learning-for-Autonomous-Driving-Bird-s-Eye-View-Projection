```python
def img_to_categorical(img, needed_labels):

  cat = np.empty((img.shape[0], img.shape[1], len(needed_labels)))

  for channel, label in enumerate(needed_labels):
    cat[:, :, channel] = np.where(np.isin(img, label), 1, 0)

  return cat

def categorical_to_img(cat):

  img = np.argmax(cat, axis=-1)
  return img

needed_labels = [0, 1, 6, 7, 8, 10, 11]
```

```python
from tensorflow.keras.utils import Sequence

class DataGenerator2D(Sequence):
    """Generates data for Keras
    Sequence based data generator. Suitable for building data generator for training and prediction.
    """
    def __init__(self, img_paths, base_path, to_fit=True, batch_size=32, shuffle=True, needed_classes = []):

        self.img_paths = img_paths.copy()
        self.base_path = base_path
        self.to_fit = to_fit
        self.batch_size = batch_size
        self.shuffle = shuffle
        self.needed_classes = needed_classes
        self.on_epoch_end()

    def __len__(self):
        """Denotes the number of batches per epoch
        :return: number of batches per epoch
        """
        return int(len(self.img_paths) // self.batch_size)

    def __getitem__(self, index):
        """Generate one batch of data
        :param index: index of the batch
        :return: X and y when fitting. X only when predicting
        """
        # Generate indexes of the batch
        current_indexes = list(range(index * self.batch_size, (index + 1) * self.batch_size))
        img_paths_temp = self.img_paths[current_indexes]

        # Generate data
        X = self._generate_X(img_paths_temp)

        if self.to_fit:
            y = self._generate_y(img_paths_temp)
            return X, y
        else:
            return X

    def on_epoch_end(self):
        """Updates indexes after each epoch
        """
        if self.shuffle == True:
            indices = np.arange(len(self.img_paths))
            np.random.shuffle(indices)
            self.img_paths = self.img_paths[indices]
            self.img_paths.reset_index(drop=True, inplace=True)


    def _generate_X(self, img_paths_temp):
        """Generates data containing batch_size images
        :param img_paths_temp: list of label ids to load
        :return: batch of images
        """
        # Initialization

        X = []

        # Generate data
        for idx, path in zip(img_paths_temp.index, img_paths_temp):
            # Store sample
            img = self._load_image(path)
            X.append(img)

        return np.array(X)

    def _generate_y(self, img_paths_temp):

        y = []

        for idx, path in zip(img_paths_temp.index, img_paths_temp):
            # Store sample
            img = self._load_image(path, front=False)
            y.append(img)
        return np.array(y)

    def _load_image(self, image_path, front=True):

        if front:
          img_dir =  '/front/'
        else:
          img_dir =  '/top/'

        img = np.load(self.base_path + img_dir + image_path)
        img = img[::2, ::2]                                          # Downsampling image by factor of 2

        if not front:
          img = img[3:(img.shape[0]//2)]
          img = np.rot90(img)
          img = img[2:-2]                                           # Crop Image to make dimensions divisible by 8
          img = img_to_categorical(img, self.needed_classes)

          return img

        img = img[2:-2, 3:-3]                                       # Crop Image to make dimensions divisible by 8
        img = img_to_categorical(img, self.needed_classes)

        return img
```

###Model (U-netxST Generator)

```python
import tensorflow as tf
from keras.layers import Input, Conv2D, UpSampling2D, BatchNormalization, LeakyReLU, Activation, concatenate, Dense, Flatten, GlobalAveragePooling2D
from keras.models import Model
from keras.layers import Layer
from keras.saving import register_keras_serializable

@register_keras_serializable()
class SpatialTransformer(Layer):
    def __init__(self, **kwargs):
        super(SpatialTransformer, self).__init__(**kwargs)

    def affine_grid(self, theta, height, width):
        x = tf.linspace(-1.0, 1.0, width)
        y = tf.linspace(-1.0, 1.0, height)
        x_t, y_t = tf.meshgrid(x, y)
        ones = tf.ones_like(x_t)
        grid = tf.stack([x_t, y_t, ones], axis=-1)
        grid = tf.reshape(grid, [-1, 3])
        return tf.matmul(grid, theta, transpose_b=True)

    def bilinear_sampler(self, img, grid):
        batch_size, height, width, channels = tf.shape(img)[0], tf.shape(img)[1], tf.shape(img)[2], tf.shape(img)[3]
        x = 0.5 * (grid[..., 0] + 1.0) * tf.cast(width - 1, tf.float32)
        y = 0.5 * (grid[..., 1] + 1.0) * tf.cast(height - 1, tf.float32)
        x0 = tf.cast(tf.floor(x), tf.int32)
        x1 = x0 + 1
        y0 = tf.cast(tf.floor(y), tf.int32)
        y1 = y0 + 1
        x0 = tf.clip_by_value(x0, 0, width - 1)
        x1 = tf.clip_by_value(x1, 0, width - 1)
        y0 = tf.clip_by_value(y0, 0, height - 1)
        y1 = tf.clip_by_value(y1, 0, height - 1)
        x0_f, x1_f, y0_f, y1_f = map(tf.cast, [x0, x1, y0, y1], [tf.float32]*4)
        batch_indices = tf.range(batch_size)[:, tf.newaxis, tf.newaxis]
        batch_indices = tf.tile(batch_indices, [1, height, width])
        indices_a = tf.stack([batch_indices, y0, x0], axis=-1)
        indices_b = tf.stack([batch_indices, y1, x0], axis=-1)
        indices_c = tf.stack([batch_indices, y0, x1], axis=-1)
        indices_d = tf.stack([batch_indices, y1, x1], axis=-1)
        Ia, Ib, Ic, Id = map(lambda idx: tf.gather_nd(img, idx), [indices_a, indices_b, indices_c, indices_d])
        wa = tf.expand_dims((x1_f - x) * (y1_f - y), axis=-1)
        wb = tf.expand_dims((x1_f - x) * (y - y0_f), axis=-1)
        wc = tf.expand_dims((x - x0_f) * (y1_f - y), axis=-1)
        wd = tf.expand_dims((x - x0_f) * (y - y0_f), axis=-1)
        return wa * Ia + wb * Ib + wc * Ic + wd * Id

    def call(self, inputs):
        img, theta = inputs
        batch_size, height, width, channels = tf.shape(img)[0], tf.shape(img)[1], tf.shape(img)[2], tf.shape(img)[3]
        grid = self.affine_grid(theta, height, width)
        grid = tf.reshape(grid, [batch_size, height, width, 2])
        return self.bilinear_sampler(img, grid)

def build_unet_generator_with_stn(needed_labels):
    input_shape = (96, 144, len(needed_labels))
    kernel_size = [8, 6, 4]
    strides = 2
    leakyrelu_alpha = 0.2
    upsampling_size = 2
    output_channels = len(needed_labels)
    input_layer = Input(shape=input_shape)
    global_features = GlobalAveragePooling2D()(input_layer)
    theta = Dense(6, activation='linear', kernel_initializer='zeros', bias_initializer='ones')(global_features)
    theta = tf.keras.layers.Reshape((2, 3))(theta)
    stn_output = SpatialTransformer()([input_layer, theta])
    encoder1 = Conv2D(64, kernel_size[0], strides=strides, padding='same')(stn_output)
    encoder1 = LeakyReLU(alpha=leakyrelu_alpha)(encoder1)
    encoder2 = Conv2D(128, kernel_size[1], strides=strides, padding='same')(encoder1)
    encoder2 = BatchNormalization()(encoder2)
    encoder2 = LeakyReLU(alpha=leakyrelu_alpha)(encoder2)
    encoder3 = Conv2D(256, kernel_size[2], strides=strides, padding='same')(encoder2)
    encoder3 = BatchNormalization()(encoder3)
    encoder3 = LeakyReLU(alpha=leakyrelu_alpha)(encoder3)
    decoder1 = UpSampling2D(size=upsampling_size)(encoder3)
    decoder1 = Conv2D(128, kernel_size[1], padding='same')(decoder1)
    decoder1 = concatenate([decoder1, encoder2])
    decoder1 = LeakyReLU(alpha=leakyrelu_alpha)(decoder1)
    decoder2 = UpSampling2D(size=upsampling_size)(decoder1)
    decoder2 = Conv2D(64, kernel_size[0], padding='same')(decoder2)
    decoder2 = concatenate([decoder2, encoder1])
    decoder2 = LeakyReLU(alpha=leakyrelu_alpha)(decoder2)
    decoder3 = UpSampling2D(size=(2, 1))(decoder2)
    output = Conv2D(output_channels, kernel_size[0], padding='same', activation='softmax')(decoder3)
    model = Model(inputs=input_layer, outputs=output)
    return model

needed_labels = list(range(7))
model = build_unet_generator_with_stn(needed_labels)
model.summary()
```

```python
import os
from keras.optimizers import Nadam
from keras.callbacks import ModelCheckpoint, TensorBoard
from tensorflow.keras.optimizers.schedules import ExponentialDecay
from time import time

# Define directories for logs and models
log_dir = f"/content/drive/MyDrive/Thesis/self_driving(UnetxSt_2)/logs/BirdsEye_{int(time())}"
model_dir = "/content/drive/MyDrive/Thesis/self_driving(UnetxSt_2)/models"

# Ensure directories exist
os.makedirs(log_dir, exist_ok=True)
os.makedirs(model_dir, exist_ok=True)

# Define ExponentialDecay learning rate schedule
lr_schedule = ExponentialDecay(
    initial_learning_rate=1e-4,  # Starting learning rate
    decay_steps=100000,          # Steps for decay
    decay_rate=0.96,             # Decay factor
    staircase=True               # Discrete (staircase) decay
)

# Nadam optimizer with the learning rate schedule
optimizer = Nadam(learning_rate=lr_schedule)

# Define the checkpoint file path
filepath = f"{model_dir}/BirdsEye_saved_model-{{epoch:02d}}-{{val_loss:.5f}}.keras"

# Define callbacks (save every epoch)
callbacks_list = [
    ModelCheckpoint(filepath, monitor='val_loss', verbose=1, save_best_only=False),  # Save model at every epoch
    TensorBoard(log_dir=log_dir)  # Log metrics for TensorBoard
]

print("Callbacks and optimizer initialized successfully.")
```

```python
@register_keras_serializable()
def dice_coef(y_true, y_pred, smooth=1):
    y_true_f = tf.reshape(y_true, [-1])
    y_pred_f = tf.reshape(y_pred, [-1])
    intersection = tf.reduce_sum(y_true_f * y_pred_f)
    return (2. * intersection + smooth) / (tf.reduce_sum(y_true_f) + tf.reduce_sum(y_pred_f) + smooth)

@register_keras_serializable()
def dice_coef_multilabel(y_true, y_pred, numLabels=len(needed_labels)):
    dice = 0
    for index in range(0, numLabels):
        dice -= dice_coef(y_true[:, :, :, index], y_pred[:, :, :, index])
    return numLabels + dice

@register_keras_serializable()
def ssim_loss(y_true, y_pred):
    return 1 - tf.reduce_mean(tf.image.ssim(y_true, y_pred, 1.0))

@register_keras_serializable()
def custom_loss(y_true, y_pred):
    return dice_coef_multilabel(y_true, y_pred) + 5 * ssim_loss(y_true, y_pred)
```
